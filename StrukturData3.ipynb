```python
#Stack dengan List
#Program Python untuk mendemonstrasikan implementasi stack menggunakan List
stack = []

#Fungsi Append untuk push elemen di stack
stack.append('a')
stack.append('b')
stack.append('c')
stack.append('d')
stack.append('e')

print('Initial Stack')
print(stack)

#Pop() untuk pop elemen dari stack di LIFO order
print('\nElements popped from stack')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after element are popped:')
print(stack)
```

```python
#Stack dengan collections.deque
#Program python untuk mendemonstrasikan implementasi stack menggunakan collection.deque
from collections import deque
stack = deque()

#Fungsi Append untuk push elemen di stack
stack.append('a')
stack.append('b')
stack.append('c')
stack.append('d')
stack.append('e')

print('Initial Stack')
print(stack)

#Pop() untuk pop elemen dari stack di LIFO order
print('\nElements popped from stack')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after element are popped:')
print(stack)
```

```python
#Stack dengan queue.LifoQueue
#Program python untuk mendemonstrasikan implementasi stack menggunakan modul queue
from queue import LifoQueue

#Initializing a stack
stack = LifoQueue(maxsize=5)

#qsize untuk menampilkan elemen di stack
print(stack.qsize())

#Fungsi Append untuk push elemen di stack
stack.put('a')
stack.put('b')
stack.put('c')
stack.put('d')
stack.put('e')

print('Full :', stack.full())
print('Size :', stack.qsize())

#Pop() untuk pop elemen dari stack di LIFO order
print('\nElements popped from stack')
print(stack.get())
print(stack.get())
print(stack.get())

print('\nEmpty :', stack.empty())
```

```python
#Stack dengan single linked list
class Node:
  def __init__(self,value):
    self.value = value
    self.next = None
class Stack:
  #Initializing a stack menggunakan dummy untuk lebih mudah
  def __init__(self):
    self.head = Node('Head')
    self.size = 0

  #Representasi string dalam stack
  def __str__(self):
    cur = self.head.next
    out = ''
    while cur:
      out +=str(cur.value)+'->'
      cur = cur.next
    return out [:-2]

  #Mendapat size dari stack
  def getSize(self):
    return self.size

  #Cek jika stack kosong
  def isEmpty(self):
    return self.size == 0

  #Ambil top item dari Stack
  def peek(self):
    if self.isEmpty():
      raise Exception('Peeking from an empty stack')
    return self.head.next.value

  #Push value ke dalam stack
  def push(self,value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  #Remove value dari stack dan return
  def pop(self):
    if self.isEmpty():
      raise Exception('Popping from an empty stack')
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

#Driver code
if __name__=='__main__':
  stack=Stack()
  for i in range (1,11):
    stack.push(i)
  print(f'Stack: {stack}')

  for _ in range (1,6):
    remove = stack.pop()
    print(f'Pop: {remove}')
  print(f'Stack: {stack}')
```
