```python
#Queue dengan List
#Program python untuk mendemonstrasikan implementasi queue menggunakan list.
queue = []
#fungsi append untuk unqueue elemen di queue
queue.append('a')
queue.append('b')
queue.append('c')
queue.append('d')
queue.append('e')

print('Initial Queue')
print(queue)

#pop (o) untuk dequeue elemen dari queue di FIFO order
print('\nElements dequeued from queue:')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue after removing elements')
print(queue)

```

```python
#Queue dengan collections.deque
from collections import deque
queue = deque()

#fungsi append untuk enqueue element di queue
queue.append('a')
queue.append('b')
queue.append('c')
queue.append('d')
queue.append('e')

print('Initial Queue')
print(queue)

#Popleft fungsi untuk dequeue element dari stack in Lifo order
print('\nElements dequeued from queue:')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('\nQueue after removing elements')
print(queue)
```

```python
#Queue dengan queue.queue
from queue import Queue

#Initializing a queue
qu = Queue(maxsize=5)
#qsize memperlihatkan nomor untuk element di queue
print(qu.qsize())
#put() fungsi untuk enqueue element di queue
qu.put('a')
qu.put('b')
qu.put('c')
qu.put('d')
qu.put('e')
print('Full:', qu.full())
print('Size:', qu.qsize())

print('\nElement dequeued from the queue')
print(qu.get())
print(qu.get())
print(qu.get())

print('\nEmpty:', qu.empty())
print('Full:', qu.full())
```

```python

#Queue dengan single linked list
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqu(self,dataval):
    if dataval not in self.queue:
      self.queue.insert(0,dataval)
      return True
    return False

  def removefromq(self):
    if len(self.queue)>0:
      return self.queue.pop()
    return ('No elements in queue!')
  def size(self):
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu('Jan'))
print(TheQueue.addtoqu('Feb'))
print(TheQueue.addtoqu('March'))
print(TheQueue.addtoqu('April'))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
