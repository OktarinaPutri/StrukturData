```python
#Binary Tree Manual
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

#Membuat Tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

#Fungsi inorder travesal
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end='')
    inorder(node.right)

print('Hasil inorder travesal dari binary tree:')
inorder(root)
```

```python
#Membuat binary search tree
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst=BinarySearchTree()
root=None
data_list=[50,30,70,20,40,60,80]
for value in data_list:
  root = bst.insert(root, value)
```

```python
#Traversal pada BST
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end = '')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end='')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end='')

print('Inorder Traversal:')
inorder(root)
print('\nPreorder Traversal:')
preorder(root)
print('\nPostorder Traversal:')
postorder(root)
```

```python
#Pencarian Nilai dalam BST
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left,key)
  return search(node.right, key)

#Uji pencarian
key = 20
result = search(root, key)
if result:
  print(f'{key} ditemukan dalam tree.')
else:
  print(f'{key} tidak ditemukan dalam tree.')

key=25
result=search(root, key)
if result:
  print(f'{key} ditemukan dalam tree.')
else:
  print(f'{key} tidak ditemukan dalam tree.')
```

```python
# Tugas 2
# PROGRAM BINARY SEARCH TREE (BST)
# Buat class node
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Buat class BST
class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Fungsi insert data ke BST
    def insert(self, root, data):

        # Kalau root kosong
        if root is None:
            return Node(data)

        # Kalau data lebih kecil masuk kiri
        if data < root.data:
            root.left = self.insert(root.left, data)

        # Kalau data lebih besar masuk kanan
        else:
            root.right = self.insert(root.right, data)

        return root


# Fungsi Travesal
# Inorder : Kiri -> Tengah -> Kanan
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

# Preorder : Tengah -> Kiri -> Kanan
def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

# Postorder : Kiri -> Kanan -> Tengah
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

# Fungsi search BST
def search(node, key):

    # Kalau node kosong / data ditemukan
    if node is None or node.data == key:
        return node

    # Kalau key lebih kecil cari ke kiri
    if key < node.data:
        return search(node.left, key)

    # Kalau key lebih besar cari ke kanan
    return search(node.right, key)

# Data Angka
npm = "002"
data_list = [
    int(npm),   # 002
    20,         # Tanggal lahir
    10,         # Bulan lahir
    79,         # ASCII huruf O
    105,        # ASCII huruf i
    30,         # Angka bebas
    40          # Angka bebas
]


#Membuat BST
bst = BinarySearchTree()
root = None

# Memasukkan semua data ke BST
for value in data_list:
    root = bst.insert(root, value)

#Output Program
print("_" * 30)
print("  PROGRAM BINARY SEARCH TREE")
print("_" * 30)

print("\nData yang dimasukkan ke BST :")
print(data_list)

print("\n" + "_" * 30)
print("    HASIL TRAVERSAL BST")
print("_" * 30)

print("\n1. Inorder Traversal")
print(" ", end=' ')
inorder(root)

print("\n\n2. Preorder Traversal")
print(" ", end=" ")
preorder(root)

print("\n\n3. Postorder Traversal")
print(" ", end=" ")
postorder(root)

# Searching BST
print("\n\n" + "_" * 30)
print("    HASIL PENCARIAN DATA")
print("_" * 30)

# Mencari dua digit terakhir NPM
key = 20

result = search(root, key)

if result:
    print(f"\nData {key} ditemukan dalam BST")
else:
    print(f"\nData {key} tidak ditemukan dalam BST")


# Mencari angka yang tidak ada dalam list
key = 9

result = search(root, key)

if result:
    print(f"Data {key} ditemukan dalam BST")
else:
    print(f"Data {key} tidak ditemukan dalam BST")
```
