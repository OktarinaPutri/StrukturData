```python
#Representasi Graf dengan Adjacency List
#Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B','C'],
    'B': ['A','D','E'],
    'C': ['A','F'],
    'D': ['B'],
    'E': ['B','F'],
    'F': ['C','E']
}

#Menampilkan graf
def print_graph(graf):
  for node in graf:
    print(f'{node} -> {graf[node]}')

print_graph(graph)
```

```python
#Traversal Graf menggunakan BFS (Breadth First Search)
from collections import deque

def bfs(graph, start):
  visited =  set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end='')
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

#Contoh pemanggilan
bfs(graph, 'D')
```

```python
#Traversal Grafh menggunakan DFS
def dfs(graph, start, visited=None):
  if visited is None:
    visited= set()
  if start not in visited:
    print(start, end='')
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

#Contoh pemanggilan
dfs(graph,'D')
```

```python
#Latihan Personal
#Lengkapi fungsi untuk menghitung jumlah tetangga dari tiap simpul dalam graf
def count_neighbors(graph):
  neighbor_counts = {}

  for node in graph:
    neighbor_counts[node] = len(graph[node])
    print(f'{node} -> {graph[node]} (Count: {len(graph[node])})')
  return neighbor_counts
counts = count_neighbors(graph)
print(f"Neighbor counts: {counts}")
```

```python
#Tugas 1
from collections import deque

# 1. Membuat graf tak berarah dengan 5 simpul dan 7 sisi
graph_baru = {
    'A': ['B', 'C', 'E'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C'],
    'E': ['A', 'C']
}

def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

# 2. Implementasi BFS
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    print("Urutan BFS:", end=" ")
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([n for n in graph[vertex] if n not in visited])
    print()

# 2. Implementasi DFS
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
        print("Urutan DFS:", end=" ")
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Eksekusi
print("Struktur Graf:")
print_graph(graph_baru)
bfs(graph_baru, 'A')
dfs(graph_baru, 'A')
print()
```

```python
#Tugas 3
def bfs_list(graph, start):
    visited = set()
    queue = deque([start])
    order = [] # List untuk menyimpan urutan
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            order.append(vertex)
            visited.add(vertex)
            queue.extend([n for n in graph[vertex] if n not in visited])
    return order

# Contoh
print("Hasil BFS (List):", bfs_list(graph_baru, 'A'))
```

```python
#Tugas 4
def find_path(graph, start, end, path=None):
    if path is None:
        path = []
    path = path + [start]

    if start == end:
        return path

    if start not in graph:
        return None

    for node in graph[start]:
        if node not in path:
            newpath = find_path(graph, node, end, path)
            if newpath: return newpath
    return None

# Contoh penggunaan:
print("Jalur A ke D:", find_path(graph_baru, 'A', 'D'))
```

```python
#Tugas 5
def is_connected(graph):
    if not graph:
        return True

    # Ambil satu simpul sembarang sebagai titik mulai
    start_node = list(graph.keys())[0]

    # Lakukan DFS untuk melihat berapa banyak simpul yang bisa dijangkau
    visited = set()
    def traverse(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                traverse(neighbor)

    traverse(start_node)

    # Jika jumlah yang dikunjungi sama dengan jumlah total simpul, maka connected
    return len(visited) == len(graph)

# Contoh penggunaan:
print("Apakah graf terhubung?", is_connected(graph_baru))
```
